```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use("bmh")
dataset = pd.read_csv("Games-Scores.csv")
print(dataset.columns)
dataset["Breakdown"] = dataset["Breakdown"].astype(str)
dataset["Rank"] = pd.to_numeric(dataset["Rank"], errors="coerce")
dataset = dataset.dropna(subset=["Breakdown", "Rank"])

x = dataset["Breakdown"]
y = dataset["Rank"]
plt.scatter(x, y, color='blue', label="Breakdown vs Rank")
plt.xlabel("Breakdown")
plt.ylabel("Rank")
plt.title("Breakdown vs. Rank")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
