```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

dataset = pd.read_csv("CausesOfDeath-Dataset.csv")

data_pie = dataset.groupby('SEX')['DISEASE'].count()  # Use count or another numeric aggregation

plt.pie(data_pie, labels=data_pie.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Pie Chart of Causes of Death by Sex')

plt.show()
```
