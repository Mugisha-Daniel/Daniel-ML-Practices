```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.use("bmh")

dataset = pd.read_csv("candy-data.csv")

plt.barh(dataset['chocolate'], dataset['fruity'], color='skyblue')

plt.title('Horizontal Bar Plot')
plt.xlabel('Fruity')
plt.ylabel('Chocolate')

plt.show()
```
