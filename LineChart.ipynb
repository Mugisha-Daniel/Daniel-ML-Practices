```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Pc-Price-Dataset.csv")

print(dataset.columns)

x = dataset["PC_Model"]
y = dataset["PC_Price"]


plt.plot(x, y, marker='o', linestyle='-', color='b', label="PC Price Trend")
plt.xlabel("PC Model")
plt.ylabel("PC Price ($)")
plt.show()
```
